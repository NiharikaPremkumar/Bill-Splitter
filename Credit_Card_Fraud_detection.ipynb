# Importing dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Loading DataSet

```python
#loading dataset to dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

# Data Pre-Processing

```python
#first five rows print
credit_card_data.head()
```

```python
credit_card_data.tail()
```

# Data Analysis

```python
#info about data
credit_card_data.info()
```

```python
#check missing values
credit_card_data.isnull().sum()
```

```python
#distribution of legit and fraudlent transaction
credit_card_data['Class'].value_counts()
```

**HIGHLY UNBALANCED DATA SET
* 0-> NORMAL
* 1-> FRAUDELENT**


```python
#SEPERATING DATA FOR analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of data
legit.Amount.describe() #amount column
#percentile not percentage(50% values are less than 22$)
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
credit_card_data.groupby('Class').mean()
#mean in all v categories
```

 Under sampling

```python
#build a sample dataset containing similar distribution of normal and fradulent transaction
#number of fradulent transaction = 492
legit_sample = legit.sample(n=492)
# takimg 492 from normal and mixing with fraudulent ones to make a perfect training set
```

concat both dataframe

```python
new_dataset= pd.concat([legit_sample,fraud],axis=0)
#axis=0 means data frame will be added one by one rowwise
#axis=1 means data frame will be added one by one columnwise
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
#value counts
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
#nature is not changed just balanced
#to check whether we got a good or bad sample
```

splitting data in features (v) and targets(class)


```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

Split data into training and testing data

```python
#model selection module used
#test size is amount to be tested; 20%
#stratify evenly distributed 0 and 1 in train and test
#random_state helps in entropy selection
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

# Model Training
**logistic regression model**

```python
model= LogisticRegression()
```

```python
#training model with training data
model.fit(x_train,y_train)
```

# Model Evaluation

```python
#accuracy score
#accuracy on training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy in training data: ",training_data_accuracy)
#printed to show over/underfitting
```

```python
#accuracy on test data
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
print("Accuracy in Testing data: ",test_data_accuracy)
```
